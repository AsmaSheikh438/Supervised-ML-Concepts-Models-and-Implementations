### Embalanced data :
When the majority of data items in your dataset represents items belonging to one class, we say the dataset is skewed or imbalanced.

* Imbalanced classification is the problem of classification when there is an unequal distribution of classes in the training dataset.
* The imbalance in the class distribution may vary, but a severe imbalance is more challenging to model and may require specialized techniques.
* Many real-world classification problems have an imbalanced class distribution, such as fraud detection, spam detection, and churn prediction.

#### We will use two techniques..
### 1. Random Under Sampling ( convert high data into low data)
We will reduce the majority of the class so that it will have same no. of as the minority.

### 2. Rnadom Over Sampling (convert low data into high data)
We will increase the size of minority is inactive class to the size of majority class is active.

```python
import pandas as pd
```

```python
dataset = pd.read_csv('purchased_data.csv')
dataset.head(5)
```

```python
# checking data is embalanced or not
dataset['Purchased'].value_counts()
```

```python
# data is embalanced 
```

```python
x= dataset.iloc[:,:-1]
y= dataset['Purchased']
```

```python
# now splitting of data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
LR.score(x_test,y_test)*100
```

```python
# check it is biased or not
LR.predict([[19,19000]])
```

```python
LR.predict([[45,26000]])
```
