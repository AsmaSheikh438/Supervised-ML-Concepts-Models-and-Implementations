### Confusion Matrics :
#### (for binary class classification)
* A confusion matrix is a simple and useful tool for understanding the performance of a classification model, like one used in machine learning or statistics.
* A confusion matrix is used to evaluate the performance of a classification model by comparing actual and predicted labels.
* It helps you evaluate how well your model is doing in categorizing things correctly.
* It also known as the error matrix.

### The Confusion Matrix Structure
Letâ€™s learn about the basic structure of a confusion matrix, using the example of identifying an email as spam or not spam.

#### True Positive (TP) - Your model predicted the positive class. For example, identifying a spam email as spam.
#### True Negative (TN) - Your model correctly predicted the negative class. For example, identifying a regular email as not spam.
#### False Positive (FP) - Your model incorrectly predicted the positive class. For example, identifying a regular email as spam.
#### False Negative (FN) - Your model incorrectly predicted the negative class. For example, identifying a spam email as a regular email.

#### ðŸ‘‰ True Positive: The number of times our actual positive values are equal to the predicted positive. You predicted a positive value, and it is correct.
#### ðŸ‘‰ False Positive: The number of times our model wrongly predicts negative values as positives. You predicted a negative value, and it is actually positive.
#### ðŸ‘‰ True Negative: The number of times our actual negative values are equal to predicted negative values. You predicted a negative value, and it is actually negative.
#### ðŸ‘‰ False Negative: The number of times our model wrongly predicts negative values as positives. You predicted a negative value, and it is actually positive.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv('placement.csv')
dataset.head(5)
```

```python
x= dataset.iloc[:,:-1]
y= dataset['placed']
```

```python
# now splitting of data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
LR.score(x_test,y_test)*100
```

```python
# not good accuracy
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
cf = confusion_matrix(y_test,LR.predict(x_test))
```

```python
sns.heatmap(cf,annot=True)
plt.show()
```

```python
precision_score(y_test,LR.predict(x_test))*100
```

```python
# not good accuracy
```

```python
recall_score(y_test,LR.predict(x_test))*100
```

```python
f1_score(y_test,LR.predict(x_test))*100
```
