### Logistic Regression (Binary classification [ SINGLE INPUT ]

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset.head(5)
```

#### We will work only on one inputs i.e, Age , and on the basis of age we will find out the purchase.

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x='Age', y='Purchased',data= dataset)
plt.show()
```

```python
# yes our data follow logistic regression
```

```python
x= dataset[['Age']]  # make it 2D
y= dataset['Purchased']
```

```python
# now splitting of data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
LR.score(x_test,y_test)*100
```

```python
# It is a good accuracy
```

```python
# let's do a prediction now
LR.predict([[40]])
```

```python
# answer is [0] ,No,  at 40 yrs that person did not purchase
```

```python
LR.predict([[45]])
```

```python
# answer is [1] means yes purchased
```

#### Now observe how the predicted line shows in a graph

```python
sns.scatterplot(x='Age', y='Purchased',data= dataset)
sns.lineplot(x='Age', y= LR.predict(x),data=dataset, color='red')
plt.show()
```
