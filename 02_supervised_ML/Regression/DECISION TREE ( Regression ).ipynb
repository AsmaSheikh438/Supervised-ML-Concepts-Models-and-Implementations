#### Aim of the Project
#### To build a Decision Tree regression model to predict an individual's salary based on their age and years of experience.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Salary_data.csv')
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
# Finding overall null values in a data
dataset.isnull().sum().sum()
```

```python
dataset['Age'].fillna(dataset['Age'].mode()[0],inplace=True)
```

```python
dataset['Years of Experience'].fillna(dataset['Years of Experience'].mode()[0],inplace=True)
```

```python
dataset['Salary'].fillna(dataset['Salary'].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
# Yes it is a Linear regression graph
```

```python
x= dataset.iloc[:,:-1]
y= dataset['Years of Experience']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

```python
dt.score(x_train,y_train)*100
```

```python
# both test and training values same so our model is not overfitted
```

```python
# 100% accuracy
```

```python
plt.figure(figsize=(50,50))
plot_tree(dt)
plt.savefig('demo.jpg')
plt.show()
```
