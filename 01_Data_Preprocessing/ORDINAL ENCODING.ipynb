### Ordinal Encoding

## Ordinal Encoding done by using two techniques
#### 1. by scikit-learn
#### 2. by map function

####  1. Ordinal Encoding by using scikit-learn

```python
import pandas as pd
```

```python
df= pd.DataFrame({'size': ['s','m','l','xl','s','m','l','s','s','l','xl','m']})
df.head(5)
```

```python
ord_data =[['s','m','l','xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[['size']])
```

```python
oe.transform(df[['size']])
```

```python
df['size_en'] = oe.transform(df[['size']])
```

```python
df
```

#### 2. Ordinal encoding by using map function

```python
ord1_data = {'s':0,'m':1,'l':2,'xl':3} # we can give any numbers by our choice
```

```python
df['size_en_map'] = df['size'].map(ord1_data)
```

```python
df
```
