### Feature Selection
A feature is an attribute has an impact on a problem or is useful for the problem, and choosing the important features for the model is known as feature selection.

* In machine learning there are many numbers of columns is present in the dataset, sometimes there are such columns which are not that important.
* Such columns which does not require much .
* Such columns are present which do not have any role in the machine learning algorithms.
* We can also merge two columns together to get a single column.
* We can remove those column which do not required.
* Feature selection is not 100% perfect technique.
* We should have knowledge of domain to perform feature seletion.

####  From a number of column of a data , we have to extract important columns and we will use that selected number of columns to make our machine learning model.

```python
import pandas as pd

from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_csv('diabetes.csv')
dataset.head(5)
```

```python
x = dataset.iloc[:,:-1]     #[:,:-1] means (:,all rows)  and  (:-1 means columns= -1)
y= dataset['Outcome']
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(Lr,k_features=8,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```

### Backward Elimination

```python
fs = SequentialFeatureSelector(Lr,k_features=8,forward=False)
fs.fit(x,y)
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
