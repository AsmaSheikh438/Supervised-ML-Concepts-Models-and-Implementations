## (NORMALIZATION) Min-Max Scaling

### Normalization: 
This method scales each feature so that all values are within the range of 0 and 1. It achieves this by subtracting the minimum value of the feature and dividing by the range (difference between maximum and minimum values).
Normalization is a fundamental technique in machine learning that transforms numerical features to a standard scale, typically between 0 and 1.

This process ensures that all features contribute equally to the analysis, preventing those with larger numerical ranges from dominating the modelâ€™s outcomes.

Understanding Normalization

Normalization is achieved through a simple formula:

```python
import pandas as pd
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('loan_test.csv')
dataset.head(5)
```

```python
# checking null values
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
# graphical nature
sns.distplot(dataset['Coapplicant_Income'])
plt.show()
```

#### when we do scaling on data ,the outliers remains same but the magnitude will get reduced

```python
# performing min-max scaling (Normalization)
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[['Coapplicant_Income']])  # It works on two dimentional (2D) data [[ ]]
```

```python
ms.transform(dataset[['Coapplicant_Income']])
```

```python
dataset['Coapplicant_Income_min'] = ms.transform(dataset[['Coapplicant_Income']])
```

```python
dataset.head(5)
```

```python
# checking dataset nature got changed or not
# Assuming `dataset` is a DataFrame and 'Coapplicant_Income' is a column in it.

plt.figure(figsize=(12, 6))

# Before Plot
plt.subplot(1, 2, 1)  
plt.title('Before')
sns.distplot(dataset['Coapplicant_Income'])

# After Plot
plt.subplot(1, 2, 2) 
plt.title('After')
sns.distplot(dataset['Coapplicant_Income_min']) 

plt.show()
```

#### here, we can see the data magnitude get reduced in between the range of 0-1
