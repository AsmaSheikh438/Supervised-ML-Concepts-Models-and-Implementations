## ENCODING:
Conversion of Categorical data into Numerical Data.

### ONE HOT ENCODING:
One-hot encoding is a technique used to convert categorical data into a binary format where each category is represented by a separate column with a 1 indicating its presence and 0s for all other categories.


## ONE HOT ENCODING & DUMMY VARIABLES

```python
import pandas as pd
```

```python
dataset = pd.read_csv('loan_test.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
# here we can see null values present in Gender column whic is a categorical data
```

```python
dataset['Gender'].fillna(dataset['Gender'].mode()[0],inplace = True)
```

```python
dataset.isnull().sum()
```

```python
# we can perform one hot encoding by get_dummies, this is a pandas method
```

```python
 en_data =  dataset['Gender']
en_data
```

```python
pd.get_dummies(en_data)
```

```python
# we have to convert our data in numerical form
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe= OneHotEncoder()
```

```python
encoded_data = ohe.fit_transform(en_data) # convert categorical data into numerical data
```
