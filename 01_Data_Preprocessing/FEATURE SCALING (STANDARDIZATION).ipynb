### What is feature scaling?
Feature scaling is the process of setting the variables on a similar scale. This is usually done using normalization, standardization, or scaling to the minimum and maximum values.

The goal is to have all variables with similar value ranges. The value range is the difference between the maximum and minimum values.



#### Effects of feature scaling on data
* After applying feature scaling, the NATURE of our data will remain same.
* After applying feature scaling, the MAGNITUDE of data (numbers) get reduced, it depends on which type of feature scaling we using.
* The pattern of our data changes after changing the types of feature scaling.
* Feature scaling best use before we going to TRAINED our data for better improvement.

```python
# using Standardization with real time example
import pandas as pd
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('loan_test.csv')
dataset.head(5)
```

```python
# checking null values
dataset.isnull().sum()
```

```python
# no null values is present in 'Applicant_Income' and 'Coapplicant_Income'
```

```python
# checking nature of dataset ie, outliers is present or not
sns.boxplot(x='Applicant_Income',data=dataset)
plt.show()
             
```

```python
dataset.describe()
```

```python
# scaling of dataset using scikit-learn
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[['Applicant_Income']]) # It works on two dimwntional (2D) data [[ ]]
```

```python
ss.transform(dataset[['Applicant_Income']])
```

```python
dataset['Applicant_Income_ss'] = pd.DataFrame(ss.transform(dataset[['Applicant_Income']]),columns=['x'])
dataset.head(5)
```

```python
dataset.describe()
```

```python
# checking dataset nature got changed or not
# Assuming `dataset` is a DataFrame and 'Applicant_Income' is a column in it.

plt.figure(figsize=(12, 6))

# Before Plot
plt.subplot(1, 2, 1)  
plt.title('Before')
sns.distplot(dataset['Applicant_Income_ss'])

# After Plot
plt.subplot(1, 2, 2) 
plt.title('After')
sns.distplot(dataset['Applicant_Income']) 

plt.show()
```

#### Nature is same but magnitude get changed
